```python
#display all row in the data

def display_all(df):
    with pd.option_context('display.max_rows', None):
        with pd.option_context('display.max_columns', None):
            return display(df)
```

### Connect each data

Based on the questions given, it is known that the data provided are related to each other. Therefore, the data will be combined using the functions in pandas for the app_train, installment_payment and prev_app data with the following code.

Merging installment_payment.csv and prev_app.csv :

```python
#remove LN_ID on installment_payment.csv
installment_payment = installment_payment.drop(columns = ["LN_ID"],
                                              axis = 1)
```

```python
#groupping installment_payment with SK_ID_PREV column, mean aggregrate

installment_payment = installment_payment.groupby("SK_ID_PREV").agg('mean').reset_index()
```

```python
#defining ins_prev as combined data between installment_payment and prev_app

ins_prev = pd.merge(prev_app, installment_payment,
                   on = "SK_ID_PREV", how = "left")
```

```python
#removing SK_ID_PREV because its unecessary now

ins_prev = ins_prev.drop(columns = ["SK_ID_PREV"], axis = 1)
```
